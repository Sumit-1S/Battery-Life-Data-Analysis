## Q5 Calculate the charge_capacity,discharge capacity for every cycle . 

##### Formula Used:

            Capacity(k+1) = Capacity(k) + (Current(k+1) * (time(k+1) - time(k))/3600) 

```python
def compute_capacity(df_phase):
    df_phase = df_phase.sort_values(['Cycle_Index', 'Date_Time']).copy()

    # Ensure Date_Time is in datetime format
    df_phase['Date_Time'] = pd.to_datetime(df_phase['Date_Time'], errors='coerce')

    # Compute time difference in hours
    df_phase['Time_diff_hr'] = df_phase.groupby('Cycle_Index')['Date_Time'].diff().dt.total_seconds().fillna(0) / 3600

    # Compute incremental capacity: I × Δt (Ah)
    df_phase['Capacity_calc'] = df_phase['Current(A)'] * df_phase['Time_diff_hr']

    # Cumulative capacity per cycle
    df_phase['Cycle_Capacity'] = df_phase.groupby('Cycle_Index')['Capacity_calc'].cumsum()

    return df_phase

charge_df = compute_capacity(charge_df)

discharge_df = compute_capacity(discharge_df)
```

### Influence of C-rate on Charge Capacity & Differences Over Cycles
Thought process:\
C-rate impacts how fast the battery is charged/discharged.

High C-rate means faster charging/discharging, which often reduces capacity due to inefficiencies and heat.

Over cycles, higher C-rates tend to accelerate degradation.

Charging and discharging phases behave differently because charging can cause polarization effects; discharging might show more consistent capacity loss.

## Q9 Use ML methods like poly. regression to extrapolate the total Charge capacity w.r.t cycle count for every c-rate.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# --- Load your dataset ---
# df = pd.read_csv('your_data.csv') # Already loaded before

# --- Filter for charge phase only ---
df_charge = df[(df['Phase'] == 'charge')]
# df_charge = df

# --- Select relevant columns ---
df_charge = df_charge[['Cycle_Index', 'C_rate', 'Capacity(Ah)']]

# --- Group by C_rate ---
df_charge['C_rate'] = df_charge['C_rate'].apply(lambda x: round(x,2))
unique_c_rates = df_charge['C_rate'].unique()

plt.figure(figsize=(14, 7))

for c_rate in unique_c_rates:
    # Subset for current c-rate
    df_sub = df_charge[df_charge['C_rate'] == c_rate]
    
    # Group by Cycle_Index to get total charge capacity per cycle
    df_grouped = df_sub.groupby('Cycle_Index')['Capacity(Ah)'].sum().reset_index()
    
    X = df_grouped['Cycle_Index'].values.reshape(-1, 1)
    y = df_grouped['Capacity(Ah)'].values

    # Fit polynomial regression model
    degree = 3
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)

    # Predict future cycles
    X_pred = np.arange(X.min(), X.max() + 20).reshape(-1, 1)  # extend 20 cycles
    y_pred = model.predict(X_pred)

    # Plot actual and predicted
    plt.plot(X, y, 'o', label=f'C-rate {c_rate:.2f} actual')
    plt.plot(X_pred, y_pred, '-', label=f'C-rate {c_rate:.2f} predicted')

plt.xlabel('Cycle Index')
plt.ylabel('Charge Capacity (Ah)')
plt.title('Polynomial Regression: Charge Capacity vs Cycle Count for each C-rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

This supports the conclusion that higher C-rates accelerate battery wear. Polynomial regression captures the underlying trend well.

## 10. Calculate State of Health (SoH) from the charge capacity of each cycle and try to visualize the degradation behaviour. 

```python
# Filter for charge phase
df_charge = df[df['Phase'] == 'charge'].copy()

# Group by Cycle to get total charge capacity per cycle
df_cycle_capacity = df_charge.groupby('Cycle_Index')['Capacity(Ah)'].sum().reset_index()

# Calculate initial capacity (usually first full cycle)
initial_capacity = df_cycle_capacity['Capacity(Ah)'].iloc[0]

# Calculate SoH as percentage of initial capacity
df_cycle_capacity['SoH (%)'] = (df_cycle_capacity['Capacity(Ah)'] / initial_capacity) * 100
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_cycle_capacity, x='Cycle_Index', y='SoH (%)', marker='o')
plt.title('State of Health (SoH) vs Cycle Index')
plt.xlabel('Cycle Index')
plt.ylabel('SoH (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The battery loses capacity as it is used, which is typical for lithium-ion batteries. The degradation is nonlinear, possibly exponential.
